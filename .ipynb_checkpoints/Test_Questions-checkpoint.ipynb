#### Drop the One column that you think will be useless for your use case

```python
# The first column just contains index values, which will be redundant using a pandas dataframe. 
df_player_data.drop(columns=df_player_data.columns[0], axis=1, inplace=True) 
```

#### Drop Rows that you think will be useless for your use case

```python
# Every row that only contains null values
df_player_data.drop(df_player_data.loc[df_player_data.isnull().all(axis=1)].index, inplace=True)
```

#### Replace null values with 0's in the '3P%' column

```python
df_player_data['3P%'] = df_player_data['3P%'].fillna(0)
```

#### Remove rows for players who Started their careers (made their Debut) before the 1980 season

```python
df_player_data.drop(df_player_data[df_player_data['Year'] < 1980].index, inplace=True)
```

#### Load in the 'all_star_appearances.pickle' file and use this data to create a column called 'all_star' that indicates whether or not a player made the All-Star team for a given year
- This file is a dictionary in which the keys are players who've made an All-Star team in their careers
- The values are all the years that the corresponding player made an All-Star team

#### Make sure this column is binary where 1 = Made All-Star team, and 0 = Did not make All-Star team

```python
with open('all_star_appearances.pickle', 'rb') as file:
    all_star_dict = pickle.load(file)

df_player_data['all_star'] = 0

for player, years in all_star_dict.items():
    df_player_data.loc[(df_player_data['Player'] == player) & (df_player_data['Year'].isin(years)), 'all_star'] = 1
```

```python
df_player_data.loc[df_player_data['all_star'] == 1]
```

#### Create a correlation map of the features

```python
# First let's transform the categorical values into numerical ones

#categories = pd.concat(map(lambda x: pd.get_dummies(df_player_data[x], drop_first=True), list(df_player_data.select_dtypes(np.object_).columns)), axis=1)
df_player_data.Player = pd.Categorical(df_player_data.Player)
df_player_data.Pos = pd.Categorical(df_player_data.Pos)
df_player_data.Tm = pd.Categorical(df_player_data.Tm)

df_player_data['player_cat'] = df_player_data.Player.cat.codes
df_player_data['pos_cat'] = df_player_data.Pos.cat.codes
df_player_data['tm_cat'] = df_player_data.Tm.cat.codes

df_player_data.head()
```

```python
# Now let's create a dataset with only clean numerical values:

df_trf = df_player_data.drop(columns=['Player', 'Pos', 'Tm'])

nan_list = df_trf.columns[df_trf.isnull().any()].tolist()

for col in nan_list:
    df_trf[col] = df_trf[col].fillna(df_trf[col].mean())

df_trf.head()
```

```python
correlation_matrix = df_trf.corr()

plt.figure(figsize=(18, 14))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Map of Features', size=20)
plt.show()
```

#### Remove Features that you consider to be too correlated or "duplicated"

```python
# Getting a list of features that are too correlated, and dropping them
high_corr_feat = correlation_matrix.columns[((correlation_matrix > 0.7) & (correlation_matrix < 1)).any()].tolist()

df_trf = df_trf.drop(columns=high_corr_feat)

df_trf.head()
```

## Part 3: Modeling

#### Build a simple classifier that predicts whether or not a player makes an All-Star team based on their stats for that season.

```python
df_features = df_trf.drop(columns='all_star')
df_target = df_trf['all_star']

x_train, x_test, y_train, y_test = train_test_split(df_features, df_target, test_size=0.3, random_state=42)
```

```python
classifier = LogisticRegression(max_iter=10000)
classifier.fit(x_train, y_train)

predictions = classifier.predict(df_features)

df_player_data['all_star_prediction'] = predictions

df_predicted_players = df_player_data[df_player_data['all_star_prediction'] == 1]

predicted_players = '; '.join(df_predicted_players.Player.values.tolist())

print(f'Predicted Players: {predicted_players}')
```

## Part 4: Data Engineering

#### Imagine that we are in 2017 and we collect this data from each of the 31 teams that play on the NBA. Each one of the teams exposes a REST API endpoint for us to extract the data and process it, this means that we have to do 31 different GET requests on each update. We are building a Web Application for ESPN and they need this data to be updated on daily basis, so when sports analysts start working in the morning they have fresh stats to go over and can give informed predictions in TV about which players will be making the All-Star team. We have an approximate of 12 users that will be using the product on daily basis.

#### At the end of each season, we will update the Pickle file (which is stored in a Bucket in Cloud Object Storage) to include the players that were included in the All-Star team and keep the dataset and model updated. All the development will be done in Python, and the architecture of the ETL process will be defined by you, along with the Cloud Service Provider (you can choose between AWS, GCP, Azure). Don't worry about the Backend and Frontend of the App, that will be defined and implemented by the Development team, but they are waiting on your decision on which database to use.

#### Would you use a dedicated server or a serverless architecture? Why?

I would choose a serverless lambda architecture because it would be more cost-effective than maintaining and managing a dedicated server that might be underutilized during off-work times. If a single job run is done at night, rather than setup up 31 separate AWS Lambda functions and events throughout the day, not only it would reduce costs even more, but also generate a more maintainable and scalable code and overall ETL process. This way, it can automatically scale up and down based on demand.

#### What Computing resources would you use? Why?

I would use AWS Lambda because it is a service that can automatically manage computing resources and run backend code in response to events like a REST request to the NBA teams' APIs, an update on the S3 Bucket files, etc. All these events and jobs would be integrated and orchestrated using Amazon EventBridge and Step Functions services.

#### What type of storage would you choose? How would you guarantee data quality and consistency?

Since the web application does not demand extensive or intensive CRUD operations, I would go with a simple Data Lake, consisting of 2 buckets in an AWS S3 Storage. In the first bucket, we would store all the raw data coming from the rest APIs, and in the second all the transformed and processed data in the form of Parquet files, and also the Pickle file. Parquet is a columnar storage file format, which means that data for each column is stored together. The storage mechanism enables better compression and typically results in smaller file sizes compared to row-based formats like CSV.

This architecture allows file versioning, which would enable checking their consistency over time. With the use of AWS Glue Catalog and Crawlers in the buckets files, it would be possible to validate data quality and integrity using Amazon Athena, with which we can query the files directly using SQL. The data in the Parquet files can also be directly accessed by the Web Application with Python and DuckDB, that allows for direct queries using SQL as well.

#### What Security considerations would you implement for these architecture?

I would consider creating different users with different permissions, which would be managed using AWS IAM. To control network traffic, I would use a network access control list - NACL - that allows or denies specific inbound or outbound traffic at the subnet level.

#### What type of monitoring, costs analysis and application performance should this architecture have?

I would use Amazon CloudWatch for monitoring and observability. It can collect monitoring and operational data in the form of logs, metrics, and events, providing a unified view of AWS resources, applications, and services that run on AWS. For cost analysis, I would use AWS Cost Explorer to visualize, understand, and manage AWS costs and usage over time, and also set up alerts if any service passes a threshold that would prevent unexpected or unwanted costs.

Workflow:

1. An event like "NBA data ready to be fetch", scheduled by EventBridge, would trigger a Lambda Function that runs the code to extract the data from the multiple APIs and load it into the first S3 bucket.
2. A second event like "Raw Data Loaded and validated" would run another lambda to transform the data in the first bucket, convert it to Parquet, and load it into the second S3 bucket.
3. An event like "End of Season" would trigger another lambda to update the Pickle file from the Parquet ones.

Notes:

- All the ETL processes and lambda functions would be integrated and orchestrated using AWS Step Functions. All the logs produced by each step would in turn be analyzed using Cloud Watch for monitoring and performance.
- After the update of the pickle file, the model could be updated using Amazon Sagemaker in case its complexity demands it.
